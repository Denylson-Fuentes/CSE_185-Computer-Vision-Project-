```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import keras as keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, AveragePooling2D

```

```python
#Getting images with contour lines
def get_contours(path):
    for entry in os.listdir(path):
        if os.path.isfile(os.path.join(path, entry)):
            
            image = os.path.join(path, entry)
            #print(image)
            img = cv2.imread(image)

            imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            ret,thresh = cv2.threshold(imgray, 110, 150, 0)

            contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

            img = cv2.drawContours(img, contours, -1, (0,255,0), 3)
            
            cv2.imwrite(image, img)
   
            
    
```

```python
Normal_path  = "./Covid19-dataset_2/train/Normal"
Covid_path  = "./Covid19-dataset_2/train/Covid"
Viral_path  = "./Covid19-dataset_2/train/Viral Pneumonia"

test_normal_path = "./Covid19-dataset_2/test/Normal"
test_covid_path = "./Covid19-dataset_2/test/Covid"
test_viral_path = "./Covid19-dataset_2/test/Viral Pneumonia"

get_contours(Normal_path)
get_contours(Covid_path)
get_contours(Viral_path)
get_contours(test_covid_path)
get_contours(test_normal_path)
get_contours(test_viral_path)
```

```python
train_datagen = ImageDataGenerator(
    rescale = 1/255,
    horizontal_flip = True,
    zoom_range = 0.3
)

test_datagen = ImageDataGenerator(
    rescale = 1/255,
    zoom_range = 0.1,
    fill_mode = "nearest"
)
```

```python
labels = ['Normal', 'Covid', 'Viral Pnuemonia']
train_dir = "./Covid19-dataset_2/train"
test_dir = "./Covid19-dataset_2/test"
train_data = train_datagen.flow_from_directory(
    train_dir,
    batch_size = 4,
    class_mode = 'categorical'
)
test_data = test_datagen.flow_from_directory(
    test_dir,
    class_mode = "categorical",
    batch_size = 2
)
```

```python
#Create the model
CNN_model = Sequential()
```

```python
#Build and Configure the model

#1st Convolution layer
CNN_model.add(Conv2D(32, kernel_size = (3,3), padding = "valid", activation = 'relu'))
CNN_model.add(Conv2D(32, kernel_size = (5,5), padding = "same"))
CNN_model.add(MaxPooling2D(pool_size = (3,3), padding = "valid"))

#2nd Convolution layer
CNN_model.add(Conv2D(64, kernel_size = (7,7), padding = "same", activation = 'relu'))
CNN_model.add(Conv2D(128, kernel_size = (3,3), padding = "valid"))
CNN_model.add(MaxPooling2D(pool_size = (3,3), padding = "valid"))

CNN_model.add(Flatten())

CNN_model.add(Dense(units = 128, activation = 'relu'))
CNN_model.add(Dense(units = 64, activation = 'relu'))
CNN_model.add(Dense(units = 16, activation = 'relu'))
CNN_model.add(Dense(units = 3, activation = 'sigmoid'))

CNN_model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
#Fit the model now
CNN_model.fit(
    train_data,
    epochs = 100,
    validation_data= test_data
)
```
